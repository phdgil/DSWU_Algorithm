# 반복문 for loop
1. for문은 리스트에 저장된 값을 자동으로 차례차례 가져옵니다.
2. 아래에 있는 코드는 for loop에 대한 예시입니다.
3. for i in zinc -> zinc안에 있는 값을 차례대로 i라는 변수에 저장한 후, print(i)가 실행될때 출력됩니다.

# 반복문에서 리스트 활용
1. url 끝에 ]n 캐릭터가 추가되어 있으면 접속시 에러가 발생합니다.
2. 그래서 일괄적으로 \n 문자를 제거한 후, 다시 리스트에 저장하는 코드를 아래 작성해봤습니다.
3. zinc_url은 텅 빈 리스트입니다. 여기에 \n 제거 과정을 한 후, 다시 url을 저장합니다.
4. for each_url in zinc: -> each_url은 zinc리스트에 있는 각각의 텍스트 (여기서는 url)을 가져와서 저장합니다.
5. each_url.strip() 사용을 하면, 문자열 앞 혹은 뒤에 붙어있는 공백을 모두 지워버립니다.
6. each_url.strip()을 통해 \n을 제거한 후, zinc_url안에 다시 저장합니다. append는 리스트에 새로운 값을 추가하는 함수입니다.

```python
# 다운받을 URL 리스트
zinc_url
```

```python
#perplexity가 작성해준 다운받는 코드입니다.

import os
import requests
import time

#으로 시작하는 문장은 무시합니다. 그래서 코드에 대한 부연 설명을 작성할 때만 사용되요. 주석(comment)라고 부릅니다.
# 저장할 폴더 (원하면 변경)
save_dir = "zinc_db"
os.makedirs(save_dir, exist_ok=True) #python에서 폴더를 직접 생성하는 방법입니다.

'''
주석이 여러 줄인 경우에는 여기에서처럼 따옴표 3개를 이어서 쓸 수 있습니다.

아래 for문 안에는 try-except 구문도 사용되었습니다.
for문의 영향을 받는 코드들은 모두 4칸 들여쓰기가 되어 있습니다.
try, except의 영향을 받는 코드들은 for문의 영향도 받기 때문에 8칸 들여쓰기가 되어 있습니다.

open은 파일을 읽을 때도 사용하지만, 파일에 내용을 추가할 때도 쓸 수 있습니다.
"wb"가 무슨 뜻인지 찾아보세요. 기본은 "r" (read)입니다.
'''
start_time = time.time()

for url in zinc_url[:10]: #[:10]은 10개까지만 테스트해보기 위해서 작성. 테스트 이후 전체 다운로드.
    filename = os.path.basename(url)  # 예: AAAA.txt
    save_path = os.path.join(save_dir, filename)
    try:
        print(f"Downloading {url} -> {save_path}")
        r = requests.get(url, stream=True)
        r.raise_for_status()
        #여기 아래는 이해가 잘 안되는 부분입니다. perplexity에 물어봤습니다.
        #다운받는 파일을 8192byte (8KB)단위로 나눠서 다운받는 방식이라고 하네요.
        with open(save_path, "wb") as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)
    except Exception as e:
        print(f"Failed to download {url}: {e}")

end_time = time.time()
```

```python
print ('다운받는데 걸린 시간:', end_time - start_time)
```
